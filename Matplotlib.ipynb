```python
import  matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline 
```

```python
x = np.linspace(1,20,30)
y = x **2
plt.plot(x,y)
```

```python
fig= plt.figure()
axes = fig.add_axes([1,1,1,1])
axes2 = fig.add_axes([0.2,0.5,0.8,0.8])
axes.plot(y,x)
axes2.plot(x,y)
```
